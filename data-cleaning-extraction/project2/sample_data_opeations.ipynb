```python
import pandas as pd
import numpy as np
dt = pd.read_csv("sample.csv")
dt
```

```python
dt.replace('-',np.NaN,inplace = True)
dt
```

```python
dt.isnull()
```

```python
dt1 = dt.dropna()
dt1
```

#First thing I am gonna do is converting the pay from dollars to rupees.

#To do that, i need to remove the dollar symbol and comma from the pay and convert
them into float as they are in object type.

```python
for i in dt1['Early Career Pay']:
    dt1.replace(i, i.replace('$',''),inplace = True)
    dt1.replace(i, i.replace(',',''),inplace = True)
dt1
```

```python
for i in dt1['Mid-Career Pay']:
    dt1.replace(i, i.replace('$',''),inplace = True)
    dt1.replace(i, i.replace(',',''),inplace = True)
dt1
```

```python
dt2 = dt1['Early Career Pay'].astype(float)
dt3 = dt1['Mid-Career Pay'].astype(float)
dt2
dt3
```

```python
dt1['Early Career Pay'] = dt2
dt1['Mid-Career Pay'] = dt3
dt1
```

```python
dt1.dtypes
```

#Now the Conversion is over

#To Convert from dollars to rupees, i need to multiply each value by 83 INR

```python
dt1['Early Career Pay'] = dt1['Early Career Pay']*82.27
dt1
```

```python
dt1['Mid-Career Pay'] = dt1['Mid-Career Pay']*82.27
dt1
```
