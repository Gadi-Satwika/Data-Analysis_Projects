```python
#So there are duplicates and null values, let's start cleaning the data
```

```python
df1 = df1.drop_duplicates()
df1
```

```python
df1.loc[:,'Age'] = df1['Age'].fillna(df1['Age'].mean())
df1.loc[:,['Gender','City']] = df1[['Gender','City']].fillna('Unknown')
df1.loc[:,'High School Percentage'] = df1['High School Percentage'].fillna(df1['High School Percentage'].mean())
```

```python
df1.isnull().sum()
```

```python
df1.loc[:,'Admission Test Score'] = df1['Admission Test Score'].astype(float)
```

```python
df1.dtypes
```

```python
df1
```

```python
df1.dropna(inplace = True)
```

```python
df1.isnull().sum()
```

```python
#Start Working with inconsistent values
```

```python
median = df1['Age'].median()
```

```python
df1.loc[ (df1['Age']<17) | (df1['Age']>24),'Age' ] = int(median)
```

```python
df1
```

```python
median_school = df1['High School Percentage'].median()
```

```python
df1.loc[ (df1['High School Percentage'] <0) | (df1['High School Percentage'] >100) , 'High School Percentage'] = median_school
```
